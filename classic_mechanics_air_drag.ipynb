```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the parameters for the problem
v0 = 1.0  # initial velocity
c1 = 0.1  # coefficient of linear drag
c2 = 0.01 # coefficient of quadratic drag

# Define the function for the velocity considering both drag forces
def velocity(v, t, c1, c2):
    return -c1*v - c2*v**2

# Create a time array for the solution
t = np.linspace(0, 10, 1000)

# Solve the differential equation
v = odeint(velocity, v0, t, args=(c1, c2)).flatten()

# Calculate the displacement by integrating the velocity
x = np.cumsum(v) * (t[1] - t[0])

# Plotting the displacement as a function of time
plt.figure(figsize=(12, 6))
plt.plot(t, x, label='Combined Drag', color='purple')

# Adding shaded regions to indicate the qualitative transition between the drag regimes
plt.axvspan(0, 2, color='yellow', alpha=0.3, label='Quadratic Dominant Region')
plt.axvspan(2, 10, color='orange', alpha=0.3, label='Linear Dominant Region')

# Adding titles and labels
plt.title('Displacement vs Time with Combined Drag Forces')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend()
plt.grid(True)
plt.savefig('plot.png', dpi=700)
plt.show()
```
